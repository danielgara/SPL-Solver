```python
%%bash
# Download concrete solver from git releases page
wget https://github.com/concrete-cp/concrete/releases/download/v3.11/concrete-3.11.txz

xz -d concrete-3.11.txz
tar xvf concrete-3.11.tar

rm -rf concrete-3.11.tar concrete-3.11.txz
```

```python
from lxml import etree
from lxml.builder import E

mx_graph_tree = etree.parse('./Example Files/mxgraph.xml')
```

```python
# Get the root
root = dict(mx_graph_tree.findall('//root')[1].items())
root = {
    'id': root['id'],
    'label': root['label']
}

root
```

```python
# Get all features for the feature model
features = []

for feature in mx_graph_tree.findall('//concrete'):
    features.append([
        dict(feature.items()),
        dict(feature[0].items()),
        dict(feature[0][0].items()),
    ])
    
features = list(filter(lambda x: not 'clon' in x[0]['id'], features))
features[:2]
```

```python
# Get all relations for the feature model
relations = []

for relation in mx_graph_tree.xpath('''//rel_concrete_root | 
                                       //rel_abstract_root | 
                                       //rel_concrete_abstract | 
                                       //rel_concrete_concrete'''):
    relations.append([
        dict(relation.items()),
        dict(relation[0].items()),
        dict(relation[0][0].items()),
    ])

relations[:2]
```

```python
# Transform features to new format
copy_features = features
features = {}

for feature in copy_features:
    features[feature[0]['id']] = {
        'label': feature[0]['label']
    }
    
features
```

```python
# Transform relations to new format
copy_relations = relations
relations = {}

for relation in copy_relations:
    relations[relation[0]['id']] = {
        'relation_type': relation[0]['relType'],
        'source': relation[1]['source'],
        'target': relation[1]['target'],
    }
    
relations
```

```python
build_rels_funcs = {
    'root': lambda x: f'eq({x}, 1)',
    'mandatory': lambda x, y: f'eq({x}, {y})',
    'optional': lambda x, y: f'ge({x}, {y})',
    'requires': lambda x, y: f'gt(add(sub(1, {x}), {y}), 0)',
    'excludes': lambda x, y: f'gt(add(sub(1, {x}), sub(1, {y})), 0)',
}
```

```python
xcsp = E.instance(
    E.variables(
        E.var(" 0 1 ", id=f"F{root['id']}", note=f"{root['label']}"),
        *[E.var(" 0 1 ", id=f'F{k}', note=f"{v['label']}") for (k, v) in features.items()]
    ),
    E.constraints(*[
        E.intension(
            E.function(build_rels_funcs[v['relation_type']](f"F{v['source']}", f"F{v['target']}"))
        ) for (k, v) in relations.items()
    ]),
    format='XCSP3', 
    type='CSP',
)
```

```python
xcsp_str = str(etree.tostring(xcsp, pretty_print=True), "UTF-8")
print(xcsp_str)
```

```python
with open("./Example Files/example_xcsp_3.xml","w+") as file:
    file.write(xcsp_str)
```
